```python
def describe_df(df,name):
    desc = df.describe(include='all').transpose().reset_index()
    missing_values = df.isnull().sum().reset_index()
    final_df = pd.merge(missing_values,desc,on='index')
    final_df.to_csv(name,index=False)
    return final_df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
```

```python
logr = LogisticRegression()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
ad = AdaBoostClassifier()
svc = SVC()
knn = KNeighborsClassifier()
```

```python
models = { 'LOGR':logr,'DT':dt, 'RF': rf,'AD':ad,'SVC':svc,'KNN':knn}
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
for key in models.keys():
    current_model = models[key]
    
    current_model.fit(smote_xtrain, smote_ytrain)
    
#     training scores
    pred_train = current_model.predict(smote_xtrain)
    
    training_recall = recall_score(smote_ytrain, pred_train)
    training_precision = precision_score(smote_ytrain, pred_train)
    train_f1_score = f1_score(smote_ytrain, pred_train)
    conf_train = confusion_matrix(smote_ytrain, pred_train)
    print(f'Model {key}:')
    print(f'Training Confusion Score: {conf_train}')
    print(f'Training Recall Score: {training_recall}')
    print(f'Training Precision Score: {training_precision}')
    print(f'Training F1 Score: {train_f1_score}')
    
    # testing scores
    
    pred_test = current_model.predict(xtest)
    
    testing_recall = recall_score(ytest, pred_test)
    test_precision = precision_score(ytest, pred_test)
    test_f1_score = f1_score(ytest, pred_test)
    conf_tst = confusion_matrix(ytest, pred_test)
    
    print(f'Model {key}:')
    print(f'Testing Confusion Score: {conf_tst}')
    print(f'Testing Recall Score: {testing_recall}')
    print(f'Testing Precision Score: {test_precision}')
    print(f'Testing F1 Score: {test_f1_score}')
    
    
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
models = {'LOGR': logr, "DT": dt, 'RF': rf, 'AD': ad, 'SVC': svc, 'KNN':knn}

params_dicts = {'LOGR': {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['lbfgs', 'liblinear', 'saga']
}, "DT": {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 10, 20, 30, 40, 50],

}, 'RF': {
    'n_estimators': [100, 200, 300],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],

}, 'AD': {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1, 10]
}, 'SVC': {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'gamma': ['scale', 'auto'],
    'degree': [3, 4, 5]  # Only relevant for 'poly' kernel
}, 'KNN':{
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'algorithm': ['ball_tree', 'kd_tree', 'brute'],
    'p': [1, 2]
}}
```

```python
best_models = {}

for key in models.keys():
    current_model = models[key]
    current_model_params = params_dicts[key]
    
    grid_search = GridSearchCV(current_model, current_model_params, scoring='recall', cv=5, verbose=2)
    
    print(f'{key}: Gridsearch Started')
    grid_search.fit(smote_xtrain, smote_ytrain)
    print(f'{key}: Gridsearch Completed')
    
    best_model = grid_search.best_estimator_
    
    best_models[key] = best_model
    
#     training scores
    pred_train = best_model.predict(smote_xtrain)
    
    training_recall = recall_score(smote_ytrain, pred_train)
    training_precision = precision_score(smote_ytrain, pred_train)
    train_f1_score = f1_score(smote_ytrain, pred_train)
    conf_train = confusion_matrix(smote_ytrain, pred_train)
    print(f'Model {key}:')
    print(f'Training Confusion Score: {conf_train}')
    print(f'Training Recall Score: {training_recall}')
    print(f'Training Precision Score: {training_precision}')
    print(f'Training F1 Score: {train_f1_score}')
    
    # testing scores
    
    pred_test = best_model.predict(xtest)
    
    testing_recall = recall_score(ytest, pred_test)
    test_precision = precision_score(ytest, pred_test)
    test_f1_score = f1_score(ytest, pred_test)
    conf_tst = confusion_matrix(ytest, pred_test)
    
    print(f'Model {key}:')
    print(f'Testing Confusion Score: {conf_tst}')
    print(f'Testing Recall Score: {testing_recall}')
    print(f'Testing Precision Score: {test_precision}')
    print(f'Testing F1 Score: {test_f1_score}')
    
    
```
